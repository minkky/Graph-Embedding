```python
import numpy as np
import glob, os
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
from keras.layers import RepeatVector
from keras.layers import TimeDistributed
from keras.preprocessing import sequence
from sklearn.model_selection import train_test_split
```

```python
dir = './sequence/*'
```

```python
# file read
all_data = []
sequence_length = []
for file in sorted(glob.glob(dir)):
    datasets = []
    for f in open(file, 'r'):
        f = f.replace(']', '').replace('[', '').replace('\n','')
        (u, v, w) = f.split(',')
        datasets.append([int(u), int(v), float(w)])
    sequence_length.append(len(datasets))
    all_data.append(datasets)
#all_data = np.array(all_data)
```

```python
max_sequence_length = max(sequence_length)
n_features = 3
batch_size = 32
epochs = 1000
```

```python
# timesteps 맞춰주기 위해서 padding 추가
all_data = sequence.pad_sequences(all_data, padding='post', maxlen = max_sequence_length)
```

```python
x_train, x_test = train_test_split(all_data, test_size = 0.2)
```

```python
model = Sequential()
model.add(LSTM(64, activation='relu', input_shape=(max_sequence_length, n_features), return_sequences=True))
model.add(LSTM(n_features, return_sequences=True))
#model.add(Dense(1))
#model.add(LSTM(n_features, return_sequences=True))
#model.add(LSTM(10, return_sequences=True))
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.fit(x_train,x_train, batch_size=batch_size, epochs=epochs, validation_data=(x_test, x_test))
```
